```python
import momma_dragonn
from momma_dragonn.data_loaders.pyfasta_data_loader import SingleStreamSeqOnly
import numpy as np
import gzip
```

```python
cell_type = "K562"
positives = cell_type+"/naive_window_around_summit.bed.gz"
fasta_data_source = "/mnt/data/annotations/by_organism/human/hg19.GRCh37/hg19.genome.fa"

data_loader = SingleStreamSeqOnly(batch_size=50, bed_source=positives, fasta_data_source=fasta_data_source,
                    rc_augment=False, randomize_after_pass=False, num_to_load_for_eval=None)
```

```python
all_data = data_loader.get_data()
```

```python
mean_X = np.mean(all_data.X,axis=0)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 15})

handles = []
handles.append(plt.plot(range(1000), mean_X[:,0]+mean_X[:,3]))
handles.append(plt.plot(range(1000), mean_X[:,1]+mean_X[:,2]))
plt.legend(handles, labels=["AT fraction", "GC fraction"])
plt.show()
```
