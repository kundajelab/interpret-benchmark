```python
from collections import OrderedDict
import gzip
import numpy as np
def load_diffs_from_file(file):
    diffs = OrderedDict()
    fp = open(file, "rb")
    for line in fp:
        line = line.decode('utf8').strip().split(": ")
        method = line[0]
        metrics = line[1].strip("[").strip("]").split(',')
        for i in range(len(metrics)):
            metrics[i] = float(metrics[i])
        diffs[method] = metrics
    fp.close()
    print("#Loaded " + str(len(diffs)) + " methods from " + file)
    return diffs
```

```python
diffs = load_diffs_from_file('no_neg_implant_diffs.txt')
methods = list(diffs.keys())
total_diffs = [d[0] for d in list(diffs.values())]
mean_diffs = [d[1] for d in list(diffs.values())]
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.tight_layout(pad=3.0)

ind = np.arange(len(methods))
width = 0.50

p1 = ax.bar(ind, total_diffs, width, bottom=0, align='center', color='green')
ax.set_title('NO IMPLANT NEG TOTAL DIFFERENCES')
ax.set_xticks(ind - width/8)
ax.set_xticklabels(methods, rotation=90)
ax.autoscale_view()

plt.show()
```

```python
fig, ax = plt.subplots()
fig.tight_layout(pad=3.0)

ind = np.arange(len(methods))
width = 0.50

p1 = ax.bar(ind, mean_diffs, width, bottom=0, align='center', color='blue')
ax.set_title('NO NEG IMPLANT MEAN DIFFERENCES')
ax.set_xticks(ind - width/8)
ax.set_xticklabels(methods, rotation=90)
ax.autoscale_view()

plt.show()
```
